```python
import pygeostat as gs
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 12}
matplotlib.rc('font', **font)
```

```python
data = gs.DataFile('sd_example.dat', x='X', y='Y', z='Z')
```

```python
data.head()
```

```python
_ = gs.location_plot(data, var='signed_distances_rt_200')
```

```python
def sqrt_coef(vertice, p):
    h = vertice[0]
    k = vertice[1]
    a = (p[1]-k)/np.sqrt(p[0]-h)
    
    return a, h, k
```

```python
def line_coef(p1, p2):
    x1, y1 = p1[0], p1[1]
    x2, y2 = p2[0], p2[1]
    a = (y2 - y1) / (x2 - x1)
    b = y1 - a * x1     
    
    return a, b
```

```python
def plt_fcd(sd, f_min, f='linear'):
    sd_min = np.min(sd)
    sd_max = np.max(sd)
    
    plt.figure(figsize=(10,3))
    plt.plot([sd_min, sd_max], [1, 1], linestyle='--', label='base case')
    if f is 'linear':
        ap, bp = line_coef([sd_min, f_min], [0, 1])
        ao, bo = line_coef([sd_max, f_min], [0, 1])
        f_maxp = ap*sd_max+bp
        f_maxo = ao*sd_min+bo
        plt.plot([sd_min, sd_max], [f_min, f_maxp], label='pessimistic')
        plt.plot([sd_min, sd_max], [f_maxo, f_min], label='optimistic')
        plt.plot([0,0],[f_min, np.max([f_maxp, f_maxo])], c='gray', linestyle='--')
        
    else:
        x_neg = np.linspace(sd_min, 0, 500)
        x_pos = np.linspace(0, sd_max, 500)
        x = np.concatenate([x_neg, x_pos])
        
        a_pos, h_pos, k_pos = sqrt_coef(vertice=[0, 1], p=[sd_max, f_min])
        a_neg, h_neg, k_neg = sqrt_coef(vertice=[0, 1], p=[np.abs(sd_min), f_min])
        y_pos = a_pos * np.sqrt(x_pos-h_pos) + k_pos
        y_neg = -a_neg * np.sqrt(np.abs(x_neg-h_neg)) + k_neg
        yo = np.concatenate([y_neg, y_pos])
        
        y_pos = -a_pos * np.sqrt(x_pos-h_pos) + k_pos
        y_neg = a_neg * np.sqrt(np.abs(x_neg-h_neg)) + k_neg
        yp = np.concatenate([y_neg, y_pos])
        
        plt.plot(x, yp, label='pessimistic')
        plt.plot(x, yo, label='optimistic')
        
        plt.plot([0,0],[np.min([yo, yp]), np.max([yo, yp])], c='gray', linestyle='--')
    
    plt.legend(fontsize=10, loc='upper left')
    plt.grid()
    plt.title('Data conditioning factor parametrization', fontsize=12)
    plt.xlabel('Signed distance', fontsize=12)
    plt.ylabel('DC factor', fontsize=12)
    plt.savefig('{}_fcd.png'.format(f), bbox_inches='tight')
    plt.show()
```

```python
plt_fcd(sd=data['signed_distances_rt_0'], f_min=0.99)
```

```python
plt_fcd(sd=data['signed_distances_rt_0'], f_min=0.99, f='quadratic')
```
